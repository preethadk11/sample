**LIST IN PYTHON**
       

```python
name=list(('apple','banana','orange'))
print(name)
print(len(name))
for i in range(len(name)):
  print(name[i])
print(name[0:3])
fruits=['apple','orange','banana','kiwi','papaya','watermelon']
print(fruits[-4:-1])
if 'apple' in fruits:
  print('yes')
fruits[2]='pears'
print(fruits)
fruits[1:3]=['watermelon']
print(fruits)
fruits.insert(1,'mango')
print(fruits)
fruits.extend(name)
print(fruits)
fruits.remove("kiwi")
print(fruits)
thislist=['apple','banana','orange']
thislist.pop(1)
print(thislist)
fruits.clear()
print(fruits)

```

**Loop list**

```python
fruits=['apple','orange','banana']
for i in fruits:
  print(i)
for i in range(len(fruits)):
  print(fruits[i])
i=0
while(i<len(fruits)):
  print(fruits[i])
  i=i+1
```

**List comprehension**

```python
thislist=['apple','orange','banana']
[print(x) for x in thislist]
fruits=['apple','banana','orange','mango']
newlist=[]
for x in fruits:
  if "a" in x:
    newlist.append(x)
print(newlist)
bag=['water','pen','eraser']
bag1=[x for x in bag if "e" in x]
print(bag1)
grocery=['apple','powder','flour']
bag=[x for x in grocery if x!='apple']
print(bag)
school=[x for x in range(10) if x<7]
print(school)
vegetables=['tomato','potato','corn']
newlist1=[x if x!='tomato' else 'brinjal'  for x in vegetables]
print(newlist1)
```
