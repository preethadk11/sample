**List comprehension**

```python
thislist=['apple','orange','banana']
[print(x) for x in thislist]
fruits=['apple','banana','orange','mango']
newlist=[]
for x in fruits:
  if "a" in x:
    newlist.append(x)
print(newlist)
bag=['water','pen','eraser']
bag1=[x for x in bag if "e" in x]
print(bag1)
grocery=['apple','powder','flour']
bag=[x for x in grocery if x!='apple']
print(bag)
school=[x for x in range(10) if x<7]
print(school)
vegetables=['tomato','potato','corn']
newlist1=[x if x!='tomato' else 'brinjal'  for x in vegetables]
print(newlist1)
```

**Nested dictionary**

```python
myfamily={
"child1":{
"name":"preetha",
"age":18,
},
"child2":{
"name":"priyanga",
"age":16,
},
"child3":{
"name":"ivan",
"age":4
}
}
for x,obj in myfamily.items():
  print(x)
  for y in obj:
    print(y+":",obj[y])
```

```python
child1={
    "name":"preetha",
     "age":18
}
child2={
    "name":"priyanga",
    "age":16
}
child3={
    "name":"ivan",
    "age":4
}
myfamily={
    "child1":child1,
    "child2":child2,
    "child3":child3
}
print(myfamily)
print(myfamily["child2"]["name"])
```
