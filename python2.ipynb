**Loop list**

```python
fruits=['apple','orange','banana']
for i in fruits:
  print(i)
for i in range(len(fruits)):
  print(fruits[i])
i=0
while(i<len(fruits)):
  print(fruits[i])
  i=i+1
```

**List comprehension**

```python
thislist=['apple','orange','banana']
[print(x) for x in thislist]
fruits=['apple','banana','orange','mango']
newlist=[]
for x in fruits:
  if "a" in x:
    newlist.append(x)
print(newlist)
bag=['water','pen','eraser']
bag1=[x for x in bag if "e" in x]
print(bag1)
grocery=['apple','powder','flour']
bag=[x for x in grocery if x!='apple']
print(bag)
school=[x for x in range(10) if x<7]
print(school)
vegetables=['tomato','potato','corn']
newlist1=[x if x!='tomato' else 'brinjal'  for x in vegetables]
print(newlist1)
```

**Functions in Python**

```python
def conversion(fahrenheit):
  return (fahrenheit-32)*(5/9)
print(conversion(95))
def greet(name):#name is parameter
  return name
print(greet("preetha"))#preetha is the argument here since it is the actual value
def number(num=0):#default value of num is 0 here
  return num
print(number(1))
print(number(2))
print(number())
def wildlife(animal,name):
  print("I have a "+animal)
  print("My "+animal+"'s names is "+name)
wildlife(name="preetha",animal="dog")
def list1(fruits):
  for x in fruits:
    print(x)
fruits=["banana","orange","apple"]
list1(fruits)
```

```python
def greet(*name):
  print("the youngest child: ",name[3])

greet("preetha","priyanga","kasthuri","doss")
def dictionary(**details):
  print("Name: ",details["age"])
print(dictionary(name="preetha",age=56))
def func(a,b,c):
  return a+b+c
num=[1,2,3]
result=func(*num)
print(result)
x=lambda a:a+4
print(x(2))
y=lambda a,b:a*b
print(y(7,8))
def func(n):
  return lambda a:a*n
doubler=func(2)
print(doubler(11))
```

```python
def countdown(n):
  if n<=0:
    print("done")
  else:
    print(n)
    countdown(n-1)
countdown(5)
def fibo(n):
  if n<=1:
    return n
  else:
    return fibo(n-1)+fibo(n-2)
print(fibo(7))
```
